```python
from jetracer.nvidia_racecar import NvidiaRacecar
car = NvidiaRacecar()
```

```python
import torch
model = torch.hub.load('./yolov3-master', 'custom', path='./yolov3-master/yolov3400.pt', source='local',force_reload=True)
```

```python
from jetcam.csi_camera import CSICamera
import cv2,time
import numpy as np
camera0 = CSICamera(width=400, height=400)
camera0.running=True
```

```python
car.throttle = -0
car.steering =0
```
